```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
df1 = pd.read_csv(mouse_drug_data_to_load)
df2 = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
df = pd.merge(df2, df1, how='left', on='Mouse ID')

# Display the data table for preview
df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint
mean_tumor_vol = df.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].mean()

# Convert to DataFrame
df_mtv = mean_tumor_vol.to_frame().reset_index()
# Preview DataFrame
df_mtv.head(15)
```

```python
# Minor Data Munging to Re-Format the Data Frames
new_df = df_mtv.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]
new_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_tumor_vol = df.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].sem()
# Convert to DataFrame
df_stv = sem_tumor_vol.to_frame().reset_index()
# Preview DataFrame
df_stv.head()
```

```python
new_df_sem = df_stv.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]
```

```python
# Generate the Plot (with Error Bars)

plt.errorbar(new_df.index, new_df.iloc[:, 0], new_df_sem.iloc[:, 0], color='green', 
             linestyle='solid', marker='o', label = new_df.columns[0], capsize = 3)
plt.errorbar(new_df.index, new_df.iloc[:, 2], new_df_sem.iloc[:, 2], color='red', 
             linestyle='solid', marker='^', label = new_df.columns[2], capsize = 3)
plt.errorbar(new_df.index, new_df.iloc[:, 3], new_df_sem.iloc[:, 3], color='red', 
             linestyle='solid', marker='s', label = new_df.columns[3], capsize = 3)
plt.errorbar(new_df.index, new_df.iloc[:, 5], new_df_sem.iloc[:, 5], color='black', 
             linestyle='solid', marker='p', label = new_df.columns[5], capsize = 3)
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Tumor Response to Treatment")
plt.legend(loc="best")
plt.grid()
# Save the Figure
plt.savefig("Images/tumor_response.png")
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met_site = df.groupby(["Drug", "Timepoint"])["Metastatic Sites"].mean()
# Convert to DataFrame
df_mms = pd.DataFrame(mean_met_site)
# Preview DataFrame
df_mms.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
mean_met_site_sem = df.groupby(["Drug", "Timepoint"])["Metastatic Sites"].sem()
# Convert to DataFrame
df_mms_sem = pd.DataFrame(mean_met_site_sem)
# Preview DataFrame
df_mms_sem.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

new_mms = df_mms.reset_index().pivot(index = "Timepoint", columns = "Drug")["Metastatic Sites"]
# Preview that Reformatting worked
new_mms
```

```python
new_mms_sem = df_mms_sem.reset_index().pivot(index = "Timepoint", columns = "Drug")["Metastatic Sites"]
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(new_mms.index, new_mms.iloc[:, 0], new_mms_sem.iloc[:, 0], color='green', 
             linestyle='solid', marker='o', label = new_df.columns[0], capsize = 3)
plt.errorbar(new_mms.index, new_mms.iloc[:, 2], new_mms_sem.iloc[:, 2], color='red', 
             linestyle='solid', marker='^', label = new_df.columns[2], capsize = 3)
plt.errorbar(new_mms.index, new_mms.iloc[:, 3], new_mms_sem.iloc[:, 3], color='red', 
             linestyle='solid', marker='s', label = new_df.columns[3], capsize = 3)
plt.errorbar(new_mms.index, new_mms.iloc[:, 5], new_mms_sem.iloc[:, 5], color='black', 
             linestyle='solid', marker='p', label = new_df.columns[5], capsize = 3)
plt.xlabel("Time (Days)")
plt.ylabel("Metastatic Sites")
plt.title("Metastasis During Treatment")
plt.legend(loc="best")
plt.grid()
# Save the Figure
plt.savefig("Images/met_sites.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count = df.groupby(["Drug", "Timepoint"])["Mouse ID"].count()
# Convert to DataFrame
mo_co_df = pd.DataFrame(mouse_count).reset_index().rename(columns={"Mouse ID": "Mouse Count"})
# Preview DataFrame
mo_co_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
new_mc = mo_co_df.pivot(index = "Timepoint", columns = "Drug")["Mouse Count"]
# Preview the Data Frame
new_mc
```

```python
# Generate the Plot (Accounting for percentages)
plt.errorbar(new_mc.index, new_mc.iloc[:, 0]/25*100, color='green', 
             linestyle='solid', marker='o', label = new_df.columns[0])
plt.errorbar(new_mc.index, new_mc.iloc[:, 2]/25*100, color='red', 
             linestyle='solid', marker='^', label = new_df.columns[2])
plt.errorbar(new_mc.index, new_mc.iloc[:, 3]/25*100, color='red', 
             linestyle='solid', marker='s', label = new_df.columns[3])
plt.errorbar(new_mc.index, new_mc.iloc[:, 5]/25*100, color='black', 
             linestyle='solid', marker='p', label = new_df.columns[5])
plt.xlabel("Time (Days)")
plt.ylabel("Survival rate (%)")
plt.title("Survival Chart")
plt.legend(loc="best")
plt.grid()
# Save the Figure
plt.savefig("Images/survival.png")
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
percent_changes = []
for i in range(len(new_df.columns)):
    percent = (new_df.iloc[9, i] - new_df.iloc[0, i]) / new_df.iloc[0, i] *100
    percent_changes.append(percent)
pc_df = pd.DataFrame(percent_changes)
pc_df.columns =["Percent Change"]
pc_df.index = new_df.columns
# Display the data to confirm
pc_df
```

```python
new_pc = pc_df.iloc[[0, 2, 3, 5], 0]
x_axis = np.arange(len(new_pc))
clr = ['green' if (x < 0) else 'red' for x in new_pc]
bar_graph = plt.bar(x_axis, new_pc, color = clr, align = "center")
plt.xticks(x_axis, new_pc.index)
plt.grid(axis="y")
plt.ylabel("Tumor Volume Change (%)")
plt.title("Tumor Change Over 45 Days")
for rect in bar_graph:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2.0, 0, '%d' % int(height) + "%", ha='center', va='bottom', fontsize=13)
plt.savefig("Images/tumor_change_percent.png")
plt.show()
```
